```python
import warnings
import sys
warnings.simplefilter('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
```

```python
url = "resources/olg_solar_weather.csv"
fiveaxis = pd.read_csv(url)
fiveaxis = fiveaxis.fillna(0)
fiveaxis.head(100)
```

```python
col = list(fiveaxis.columns)
# print(col)
```

```python
print(col)
```

```python
# col = ['power_delivered', 'energy_delivered', 'cumulative_energy', 'clouds_all', 'temp_f', 'rain_1h', 'snow_1h', 'hour', 'day', 'month']
```

```python
fiveaxis.describe()
```

```python
relationships = []
for i in col:
    try:
        for j in col:
            try:
                current = i
                newlist = j
                X = fiveaxis[newlist].values.reshape(-1,1)
                y = fiveaxis[current].values.reshape(-1,1)
                X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
                model = LogisticRegression()
                model.fit(X_train, y_train)
                print(f"{current}, {newlist} train: {model.score(X_train, y_train)}, test: {model.score(X_test, y_test)}")
            except ValueError:
                print(f"skip in j dimension {j}")
    except ValueError:
        print(f"skip in i dimension {i}")
```
